* Ortalama yıllık maaşı en yüksek olan meslek (```Occupation```) nedir? ```Data Scientist``` kaçıncı sıradadır? **Not:** Aralık olarak verilen maaşların orta noktasını maaş değeri olarak alıp ```ConvertedComp``` isimli yeni bir sütun ekleyiniz.
* ```Data Scientist``` olarak çalışanların ülke bazında ortalama yıllık maaşlarını büyükten küçüğe sıralayın, ilk 10 ülkeyi ve bu ülkelerdeki ortalama yıllık ```Data Scientist``` maaşlarını listeleyin.
* 250’den fazla kişi çalışan kurumdakilerin ortalama maaşları ne kadardır?
* Hesaplanan maaş (```ConvertedComp```) sütununu kullanarak ülke bazlı ortalama, standart sapma ve medyan değerlerinden oluşan bir dataframe oluşturun ve listeleyin. Türkiye kaçıncı sıradadır ve değeri kaçtır?

````python
survey2 = survey.reset_index()
````

````python
survey2['Q24']
````

````python
survey2['Q24'].unique()
````

````python
survey2['ConvertedComp'] = survey2['Q24']
survey2['ConvertedComp'] = survey2['ConvertedComp'].astype(str).map(lambda x: x.lstrip('>'))
survey2['ConvertedComp'] = survey2['ConvertedComp'].astype(str).map(lambda x: x.lstrip())
survey2['ConvertedComp'] = survey2['ConvertedComp'].astype(str).map(lambda x: x.lstrip('$'))
survey2['ConvertedComp'] = survey2['ConvertedComp'].astype(str).map(lambda x: x.replace(',',""))
````

````python
survey2['ConvertedComp'].unique()
````

````python
survey2['ConvertedComp'] = survey2['ConvertedComp'].astype(str).map(lambda x: " ".join(x.split("-")))
survey2['ConvertedComp']
````

````python
from tqdm import tqdm  
for i in tqdm(range(0,len(survey2['ConvertedComp']))):
    if survey2['ConvertedComp'][i] != 'nan':
        if " " in survey2['ConvertedComp'][i]:
            t = survey2['ConvertedComp'][i].split(" ")
            survey2['ConvertedComp'][i] = (int(t[0])+int(t[1]))/2
        else:
            survey2['ConvertedComp'][i] = int(survey2['ConvertedComp'][i])
````

````python
survey2['ConvertedComp']
````

````python
import numpy as np
df_8_1 = survey2[survey2['ConvertedComp']!='nan'].groupby('Occupation')['ConvertedComp'].apply(np.mean).reset_index()
df_8_1.sort_values(by='ConvertedComp', ascending=False)
#data science görüldüğü üzere 2.sıradadır.
````

## Soru 13:
* https://download.geonames.org/export/dump/countryInfo.txt adresindeki veriyi:
    * ilk 49 satırını ihmal ederek
    * yalnızca ```#ISO, Country, neighbours``` sütunlarını kullanarak 
    * ```Country``` sütununu indeks sütun olarak  
    
okuyarak __country_info__ adlı DataFrame'e aktarınız.
* ```#ISO``` sütununun ismini ```Country Code``` olarak değiştiriniz.
* ```survey``` ve ```country_info``` adlı DataFrame'leri ülke bazında birleştirip ```merged_survey``` adında yeni bir dataframe'e aktarınız.

````python
country_info = pd.read_csv('countryInfo.txt', "\t", skiprows = [i for i in range(0, 49)], usecols=["#ISO", "Country", "neighbours"], index_col= 'Country')
country_info
````

````python
country_info.rename(columns={'#ISO':'Country Code'}, inplace=True)
country_info
````

Almanya ve komşu ülkeleri yeni gelen datada bulunduğu için inner join kullandım. 5000 satır az geliyor left joine göre. Left join de kullanılabilirdi.

````python
merged_survey = survey.merge(country_info, how='inner', on='Country')
merged_survey
````

pycaret kodun sonundaki model explanability görseli için built-in bir şekilde shap kütüphanesini kullanıyor.

````python
from pycaret.regression import *
````

````python
df_bonus = survey2.loc[(survey2['Country']=='United States of America') & (survey2['Occupation']=='Data Scientist')].filter(regex='^Q16', axis=1)        
df_bonus['Age'] = survey2.Age
df_bonus['Salary(mean)'] = survey2.ConvertedComp
df_bonus = df_bonus[df_bonus['Salary(mean)'] != 'nan']
df_bonus = df_bonus.replace(np.nan, 0)
df_bonus
````

````python
df_bonus.info()
````

````python
reg = setup(data = df_bonus, 
             target = 'Salary(mean)', train_size = 0.8, fold=10,
             categorical_features = list(df_bonus.columns[:-1].values), silent = True)
````

````python
compare_models() 
# En başarılı sonuç huber reggressor fakat model explanability açısından istediğim çıktıyı almak için(tree-based gerekiyor)
# rf regressor kullanacağım
````

````python
rf_reg = create_model('rf')
````

````python
tuned_rf = tune_model(rf_reg)
````

````python
evaluate_model(tuned_rf)
````

````python
plot_model(estimator = rf_reg, plot = 'learning') # hyperparameter tuning öncesi learning curve
````

````python
plot_model(estimator = tuned_rf, plot = 'learning') # hyperparameter tuning sonrası learning curve
````

````python
interpret_model(tuned_rf) #Pycaret kütüphanesi kendi içinde shap kütüphanesini kullanıyor
````
